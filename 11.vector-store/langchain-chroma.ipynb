Using Chroma as a Vector Store in Langchain =================================================
In this notebook, we will learn how to use Chroma as a vector store in Langchain. We will create a Chroma vector store, add some documents to it, and then perform a similarity search.
First, we need to import the necessary libraries and create a Chroma vector store.


```python
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv
load_dotenv(override=True)

# embedding = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")   we can use this one as well
embedding = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview")
```

```python
from langchain_core.documents import Document

# Create a Chroma vector store and add documents to it 
# We are creating a Chroma vector store with the specified embedding function, persist directory, and collection name. Then we are adding a list of documents to the vector store.

doc1 = Document(
    page_content="Virat Kohli is an Indian cricketer and former captain of the Indian national team.",
    metadata={"team":"Royal Challengers Bangalore"}
)
doc2 = Document(
    page_content="MS Dhoni is an Indian cricketer and former captain of the Indian  national team.",
    metadata={"team":"Chennai Super Kings"}
)
doc3 = Document(
    page_content="Rohit Sharma is an Indian cricketer and current captain of the Indian national team.",
    metadata={"team":"Mumbai Indians"}
)
docs=[doc1, doc2, doc3]
```

```python
# Create a Chroma vector store and add documents to it
vector_store = Chroma(
    embedding_function=embedding,
    persist_directory="./chroma_db",
    collection_name="cricket_players"
)
```

```python
# Add documents to the vector store
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings', 'documents', 'metadatas'])
```

```python
#Search for similar documents k means how many similar documents we want to retrieve
#query means what we want to search for in the vector store

vector_store.similarity_search(
    query="Who is the good bowler?",
    k=2
)
```

```python
#similarity score showing result with score how much it is similar to the query

vector_store.similarity_search_with_score(
    query="Who is the good amazing batman?",
    k=2
)
```

```python
vector_store.similarity_search_with_score(
    query="Tell me about Royal Challengers Bangalore players",
    filter = {"team":"Royal Challengers Bangalore"},
    k=1
)
```
