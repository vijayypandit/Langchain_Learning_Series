```python
from langchain_classic.retrievers.multi_query import MultiQueryRetriever
from langchain_core.documents import Document
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv(override=True)
```

```python
# Create documents
all_docs = [
    Document(
        page_content="* Drink at least 8 glasses of water daily to maintain organ function.\n* Carry a reusable water bottle to track intake.\n* Replace sugary sodas with sparkling water or herbal infusions.",
        metadata={"H1": "Hydration Essentials"}
    ),
    Document(
        page_content="* Fill half your plate with colorful vegetables and fruits.\n* Prioritize whole grains like quinoa and brown rice over refined carbs.\n* Include lean proteins such as beans, lentils, or fish in every meal.",
        metadata={"H2": "Balanced Nutrition"}
    ),
    Document(
        page_content="* Chew slowly to allow your brain to register fullness signals.\n* Avoid screens and distractions during mealtime.\n* Listen to hunger cues rather than eating out of boredom.",
        metadata={"H3": "Mindful Eating Practices"}
    ),
    Document(
        page_content="* Aim for 30 minutes of moderate aerobic activity five days a week.\n* Incorporate strength training to protect bone density.\n* Take short walking breaks every hour if working at a desk.",
        metadata={"H4": "Physical Activity Routine"}
    ),
    Document(
        page_content="* Solar system is a collection of planets, moons, asteroids, comets, and other celestial bodies that orbit around the Sun. It includes eight planets: Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, and Neptune. The solar system also contains dwarf planets like Pluto, as well as countless smaller objects such as asteroids and comets. The Sun is the central star of the solar system and provides the necessary light and heat to sustain life on Earth.",
        metadata={"H5": "Sleep Hygiene"}
    ),
    Document(
        page_content="* Practice deep breathing exercises or meditation daily.\n* Spend time outdoors in nature to lower cortisol levels.\n* Set boundaries between work and personal life to prevent burnout.",
        metadata={"I1": "Stress Management"}
    ),
    Document(
        page_content="* Choose raw nuts and seeds for healthy fats and protein.\n* Keep pre-cut vegetables like carrots and peppers ready for quick snacks.\n* Opt for Greek yogurt with berries instead of processed sweets.",
        metadata={"I2": "Healthy Snacking"}
    ),
    Document(
        page_content="* Read nutrition labels to identify hidden sugars and high sodium.\n* Cook meals at home using fresh ingredients whenever possible.\n* Avoid foods with long lists of artificial preservatives.",
        metadata={"I3": "Reducing Processed Foods"}
    ),
    Document(
        page_content="* Brush twice daily and floss to prevent systemic inflammation.\n* Wash hands frequently to reduce the spread of germs.\n* Use sunscreen daily to protect skin from UV damage.",
        metadata={"I4": "Oral and Personal Hygiene"}
    ),
    Document(
        page_content="* Focus on small, sustainable changes rather than fad diets.\n* Track your progress in a journal to stay motivated.\n* Forgive yourself for slip-ups and restart at the next meal.",
        metadata={"I5": "Consistency and Mindset"}
    )
]
```

```python
embedding = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview") 

vectorstore = FAISS.from_documents(documents=all_docs,embedding=embedding)
```

```python
#Create retriever
similarity_retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 5})
```

```python
#Multi-query retriever
multiquery_retriever = MultiQueryRetriever.from_llm(
    retriever=vectorstore.as_retriever(search_kwargs={"k": 5}),
    llm=ChatGroq(model="llama-3.1-8b-instant")

)
```

```python
#Query
query = "How to improve energy levels and reduce stress?"
#Rerieve relevant documents
similarity_result = similarity_retriever.invoke(query)
multiquery_result = multiquery_retriever.invoke(query)

for i ,doc in enumerate(similarity_result):
    print(f"Similarity Retriever Result {i+1}: {doc.page_content}\n")

for i,doc in enumerate(multiquery_result):
    print(f"Multi-query Retriever Result {i+1}: {doc.page_content}\n")
```
