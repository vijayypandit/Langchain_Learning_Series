```python
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_core.documents import Document

from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
#Step1 create document
documents = [
    Document(page_content="Langchain helps to build LLM applications easily."),
    Document(page_content="Langchain provides various modules to build LLM applications."),
    Document(page_content="Embeddings are vector representations of text that capture semantic meaning."),
    Document(page_content="Chroma is a vector store that can be used to store and retrieve embeddings.")
]
```

```python
#step 2 
embedding = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview") 

#step3 Create a vector store and add documents
vector_store = Chroma.from_documents(documents, embedding,collection_name="langchain_docs")
```

```python
#step4 Create a retriever from the vector store
retriever = vector_store.as_retriever(search_kwargs={"k": 2})
```

```python
query = "what is chroma db used for? "
results = retriever.invoke(query)
results
```

```python
#enable MMR in the retriever 
#MMR stands for Maximal marginal relevance, which is a technique used to diversify the results returned by a retriever. It helps to ensure that the retrieved documents are not only relevant to the query but also diverse in terms of content.
#Create a retriever with MMR enabled

retriever = vector_store.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 4, "lambda_mult": 0.5}
)

query = "what is langchain used for? "
results = retriever.invoke(query)
results

for i ,docs in enumerate(results):
    print(f"\n---Result {i+1}---")
    print(f"Document {i+1}:\n{docs.page_content}\n")
```
