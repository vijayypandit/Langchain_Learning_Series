```python
from langchain_core.messages import SystemMessage, HumanMessage , AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from dotenv import load_dotenv
load_dotenv(override=True)
import os
#Dynamic prompts with ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ('system', 'You are a helpful {domain} expert'),
    ('human', 'Explain in simple terms, what is {topic}')
])

prompt = chat_template.invoke({"domain": "AI", "topic": "machine learning"})
print(prompt)
```

##Message Placeholder in ChatPromptTemplate for Dynamic Prompting

```python
from langchain_core.prompts import ChatPromptTemplate , MessagesPlaceholder
chat_template = ChatPromptTemplate([
    ('system', 'You are a customer support agent.'),
    MessagesPlaceholder(variable_name="chat_history"),
    ('human', '{query}')
])
chat_history = []

#load chat history
with open('chat_history.txt') as f:
    chat_history.extend(f.readlines())


print(chat_history)

#Create prompt
response_prompt = chat_template.invoke({'chat_history':chat_history, 'query': HumanMessage(content='What is the status of my order?')})
print(response_prompt)
```
