Pydantic Implementation :  Use for data validation and settings management using Python type annotations. It provides a way to define data models with type hints, and it automatically validates and parses the data based on those annotations. Pydantic is commonly used in FastAPI for request and response validation, as well as in other applications where data integrity is important.

```python
from langchain_groq import ChatGroq
from langchain_google_genai import ChatGoogleGenerativeAI

import streamlit as st
import os
from dotenv import load_dotenv
load_dotenv(override=True)

# model=ChatGroq(model="llama-3.1-8b-instant")
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
from pydantic import BaseModel, EmailStr, Field

class Student(BaseModel):
    name: str 


new_student = {'name':'32'}
student = Student(**new_student)
print(new_student)
```

```python
class Student(BaseModel):
    name: str  = 'vijay'


new_student = {}
student = Student(**new_student)
print(student.name)
```

```python
from typing import Optional
class Student(BaseModel):
    name: str  = 'vijay'
    age: Optional[int] = None
    email: Optional[EmailStr] = None

new_student = {'age' : 29, 'email': 'abc@email.com'}
student = Student(**new_student)
print(student)
```
