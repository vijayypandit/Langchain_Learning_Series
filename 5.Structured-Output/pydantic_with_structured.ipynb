Using Pydantic to get Structured Output from LLMs
With validation and parsing capabilities, Pydantic models can be used to define the expected structure of the output from language models. This allows for more reliable and consistent results when working with LLMs, as the output can be validated against a predefined schema.
More powerful than TypedDict, Pydantic models can also include validation rules, default values, and more complex data types, making them a robust choice for structured output from LLMs.
============================================== 

```python

class Review(BaseModel):

    key_themes: list[str] = Field(description="Write down all the key themes discussed in the review")
    summary: str = Field(description="A brief summary of the review")
    sentiment: Literal["neg", "pos", "neutral"] = Field(description="Return the sentiment of the review either ng,pos,neutral")
    pros: Optional[list[str]] = Field(default=None, description="Write down all pros in List ")
    cons: Optional[list[str]] = Field(default=None, description="Write down all cons in List ")
    name: Optional[str] = Field(default=None, description="Write the name of Reviewer from the review")

model=ChatGroq(model="llama-3.1-8b-instant")

structured_model = model.with_structured_output(Review)

result_with_structure=structured_model.invoke("""
As i recentlly upgraded to The Samsung Galaxy S24 FE delivers excellent value with its vibrant 6.4-inch display and smooth 120Hz refresh rate, making everything feel responsive and enjoyable. 

pros:
- Vibrant 6.4-inch display with smooth 120Hz refresh rate
- Impressive camera system with good daylight photos and strong night mode performance
- Full-day battery life with convenient fast charging capabilities
- Premium design with metal frame and quality build
- Excellent value for mid-range pricing
- Responsive performance for daily tasks

cons:
- No microSD card slot for storage expansion
- 5G performance not significantly faster than 4G in many regions
- Bloated software with pre-installed Samsung apps
- Occasional stuttering with demanding games
- Plastic back feels less premium than flagship models
- Water resistance limited to IP54 instead of IP68
- Performance hiccups with intensive multitasking

Review by Vijay 
""")

```
