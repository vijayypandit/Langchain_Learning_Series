#Strcutured Output Using Langchain Models -----------------------------

```python
from typing import TypedDict

class Person(TypedDict):
    name: str
    age: int

new_Person: Person = {
    "name": "Alice", "age": 30
}
print(new_Person)
```

```python
from langchain_groq import ChatGroq
from langchain_google_genai import ChatGoogleGenerativeAI

import streamlit as st
import os
from dotenv import load_dotenv
load_dotenv(override=True)

# model=ChatGroq(model="llama-3.1-8b-instant")
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

#Schema for structured output

class Review(TypedDict):
    summary: str
    sentiment: str

structured_model = model.with_structured_output(Review)
```

```python
result=structured_model.invoke("""
The hardware is solid and the design is sleek, making it a pleasure to use.
There are some minor issues with the software, but overall, it's a great product for the price.
""")
# print(result)
print(result["summary"])
print(result["sentiment"])
```

Annotated Output using TypedDict ,Annotated,Optinal,Literal

```python
from typing import  TypedDict,Annotated,Optional

class Review1(TypedDict):

    key_themes:Annotated[list[str], "Write down all key themes discussed in the review"]
    summary: Annotated[str, "A brief summary of the review"]
    sentiment: Annotated[str, "Return the sentiment of the review: neg, pos, neutral"]
    pros: Annotated[Optional[list[str]], "Write down all pros in List "]
    cons: Annotated[Optional[list[str]], "Write down all cons in List "]

# model=ChatGroq(model="llama-3.1-8b-instant")

structured_model1 = model.with_structured_output(Review1)

result=structured_model1.invoke("""
The Samsung Galaxy S24 FE delivers excellent value with its vibrant 6.4-inch display and smooth 120Hz refresh rate, making everything feel responsive and enjoyable. 

PROS:
- Vibrant 6.4-inch display with smooth 120Hz refresh rate
- Impressive camera system with good daylight photos and strong night mode performance
- Full-day battery life with convenient fast charging capabilities
- Premium design with metal frame and quality build
- Excellent value for mid-range pricing
- Responsive performance for daily tasks

CONS:
- No microSD card slot for storage expansion
- 5G performance not significantly faster than 4G in many regions
- Bloated software with pre-installed Samsung apps
- Occasional stuttering with demanding games
- Plastic back feels less premium than flagship models
- Water resistance limited to IP54 instead of IP68
- Performance hiccups with intensive multitasking

Overall, the S24 FE offers tremendous value for mid-range smartphone users seeking flagship features at an affordable price. It's a solid choice for anyone upgrading from an older phone or seeking a reliable daily driver despite minor drawbacks.

""")
# print(result)
print(result["summary"])
```

```python
from typing import  TypedDict,Annotated,Optional,Literal

class Review2(TypedDict):

    key_themes: Annotated[list[str], "Write down all key themes discussed in the review"]
    summary: Annotated[str, "A brief summary of the review"]
    sentiment: Annotated[Literal["neg", "pos", "neutral"], "Return the sentiment of the review"]
    pros: Annotated[Optional[list[str]], "Write down all pros in List "]
    cons: Annotated[Optional[list[str]], "Write down all cons in List "]
    name: Annotated[Optional[str],"Write name of reiewer if mentioned in the review"]

# model=ChatGroq(model="llama-3.1-8b-instant")

structured_model2 = model.with_structured_output(Review2)

result_with_structure=structured_model2.invoke("""
As i recentlly upgraded to The Samsung Galaxy S24 FE delivers excellent value with its vibrant 6.4-inch display and smooth 120Hz refresh rate, making everything feel responsive and enjoyable. 

pros:
- Vibrant 6.4-inch display with smooth 120Hz refresh rate
- Impressive camera system with good daylight photos and strong night mode performance
- Full-day battery life with convenient fast charging capabilities
- Premium design with metal frame and quality build
- Excellent value for mid-range pricing
- Responsive performance for daily tasks

cons:
- No microSD card slot for storage expansion
- 5G performance not significantly faster than 4G in many regions
- Bloated software with pre-installed Samsung apps
- Occasional stuttering with demanding games
- Plastic back feels less premium than flagship models
- Water resistance limited to IP54 instead of IP68
- Performance hiccups with intensive multitasking
""")
print(result_with_structure)
print("Summary: ",result_with_structure['summary'])
print("Sentiment: ", result_with_structure['sentiment'])
```
