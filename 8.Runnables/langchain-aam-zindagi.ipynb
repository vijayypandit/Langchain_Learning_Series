```python
import random
# This comment was added to the notebook cell

class NakliLLM:
    def __init__(self):
        print("LLM Created")

    def predict(self,prompt):

        response_List = [
            'Delhi is capital of India',
            'Mumbai is bollywood city of India',
            'Bangalore is IT capital of India'
        ]

        return {'response':random.choice(response_List)}
```

```python
llm = NakliLLM()
```

```python
llm.predict("What is the capital of India?")
```

```python
class NakliPromptTemplate:
    def __init__(self,template,input_variables):
        self.template = template
        self.input_variables = input_variables
    
    def format(self,input_dict):
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate("Write a {length} poem about {topic}",
    input_variables=["length", "topic"]
    )       
```

```python
prompt = template.format({"length":"short","topic":"nature"})
```

```python
llm = NakliLLM()
```

```python
llm.predict("prompt")
```

```python
class NakliLLMChain:
    def __init__(self,llm,prompt):
        self.llm = llm
        self.prompt = prompt
    
    def run(self,input_dict):
        fianl_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(fianl_prompt)

        return result['response']
```

```python
template = NakliPromptTemplate("Write a {length} poem about {topic}",
    input_variables=["length", "topic"]
    )       
```

```python
llm = NakliLLM()
```

```python
chain = NakliLLMChain(llm,template)
```

```python
chain.run({'length':'short','topic':'nepal'})
```
