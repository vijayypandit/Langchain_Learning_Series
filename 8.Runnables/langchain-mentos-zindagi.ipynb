```python
from abc import ABC, abstractmethod

class Runnable(ABC):
    @abstractmethod
    def invoke(input_data):
        pass
```

```python
import random
# This comment was added to the notebook cell

class NakliLLM(Runnable):
    def __init__(self):
        print("LLM Created")

    def invoke(self,prompt):
        response_List = [
            'Delhi is capital of India',
            'Mumbai is bollywood city of India',
            'Bangalore is IT capital of India'
        ]

        return {'response':random.choice(response_List)}

    def predict(self,prompt):

        response_List = [
            'Delhi is capital of India',
            'Mumbai is bollywood city of India',
            'Bangalore is IT capital of India'
        ]

        return {'response':random.choice(response_List)}

    
```

```python
llm = NakliLLM()
```

```python
class NakliPromptTemplate(Runnable):
    def __init__(self,template,input_variables):
        self.template = template
        self.input_variables = input_variables
    
    def invoke(self,input_dict):
        return self.template.format(**input_dict)
    
    def format(self,input_dict):
        return self.template.format(**input_dict)

    
```

```python
class NakliStrOutputParser(Runnable):
    def __init__(self):
        pass
    
    def invoke(self,input_data):
        return input_data['response']
    
```

```python
class RunnableConnector(Runnable):
    def __init__(self, runnable_list):
        self.runnable_list = runnable_list
    
    def invoke(self, input_data):
        result = input_data
        for runnable in self.runnable_list:
            result = runnable.invoke(result)
        return result
```

```python
template = NakliPromptTemplate("Write a {length} poem about {topic}",
    input_variables=["length", "topic"]
    )       
```

```python
lm = NakliLLM()
```

```python
parser = NakliStrOutputParser()
```

```python
chain  = RunnableConnector([template,llm, parser])
chain.invoke({'length':'20 line','topic':'NEPAL'})
```

```python
template1 = NakliPromptTemplate("Write a joke about {topic}",
    input_variables=["topic"]
)
```

```python
template2 = NakliPromptTemplate('Explain the following joke {response} ',
    input_variables=['response']

)
```

```python
llm = NakliLLM()
parser = NakliStrOutputParser()
```

```python
##CHAIN 1  and then CHAIN 2  --- Total Two Chaining

chain1 = RunnableConnector([template1,llm])
chain2 = RunnableConnector([template2,llm,parser])

final_chain = RunnableConnector([chain1,chain2])
final_chain.invoke({'topic':'programming'})  
```
