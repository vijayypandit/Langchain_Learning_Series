```python
from langchain_groq import ChatGroq
from langchain_core.tools import tool
import os 
from dotenv import load_dotenv
import requests

# Load environment variables from .env file
load_dotenv()

# Retrieve API key from environment
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")


# Initialize Groq Chat Model
llm = ChatGroq(
    model="llama-3.1-8b-instant"
)
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
# Initialize DuckDuckGo Search Tool
search_tool = DuckDuckGoSearchRun()
results = search_tool.run("What is the capital of France?")
print(results)
```

```python
from langchain_classic import hub
# from langchain.agents import  create_agent
from langchain_classic.agents import AgentExecutor , create_react_agent

# Load the agent from the hub 
```

```python
# Step 2: pull the react agent from the hub
# NOTE: This repository path must exist in LangChain Hub / LangSmith.
# A 404 means the repo or latest commit is not available for the given owner/repo.
try:
    prompt = hub.pull("hwchase17/react")
except Exception as exc:
    print("Failed to pull from LangChain Hub. Check the repository path and hub access.")
    print("Repo tried: hwchase17/react")
    print("Error:", exc)
    # Replace with a valid repo path or create the prompt locally.
    # Example: prompt = hub.pull("langchain/hello-world")
    raise
```

```python
#Step 3 : create the React Agent Manually with the pulle prompt
agent = create_react_agent(
    llm=llm,
    tools=[search_tool],
    prompt=prompt)
agent
```

```python
#Step 4: agent Exxecutor wrapping with default prompt and tools
agent_executor = AgentExecutor(
    agent=agent,
     tools=[search_tool], 
     verbose=True)


agent_executor 
```

```python
#Step 5 : Invoking 
response = agent_executor.invoke({"input":"what are the three ways to reach goa from Delhi ?"})
print(response)

```
