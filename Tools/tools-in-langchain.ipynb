```python
#Step 2: add type int
def multiply(a:int,b:int) -> int:
    """multiply two numbers"""
    return a*b
```

```python
from langchain_core.tools import tool
```

```python
#Step3: add tool decorator to make it a tool

@tool
def multiply(a:int,b:int) -> int:
    """multiply two numbers"""
    return a*b
```

```python
result = multiply.invoke({"a":5,"b":6})
print(result)
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
#Method 2 :Structured Tool 

from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(..., description="The first number to multiply")
    b: int = Field(..., description="The second number to multiply")
```

```python
def multiply_fun(a: int, b: int) -> int:
    """multiply two numbers"""
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply_fun,
    name = "multiply",
    description = "A tool to multiply two numbers",
    args_schema = MultiplyInput
)
```

```python
result = multiply.invoke({'a':5,'b':6}) 
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

BASE TOOL :: It is the abstract class which is used to create tools in langchain. It has two main methods which are invoke and args. Invoke method is used to call the tool and args method is used to get the arguments of the tool.

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to multiply")
    b: int = Field(required=True, description="The second number to multiply")
```

```python
from langchain_core.tools import BaseTool
from typing import Type

class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "A tool to multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
# Diagnostic: inspect MultiplyInput and BaseModel
import inspect
print('BaseModel:', BaseModel)
print('MultiplyInput bases:', MultiplyInput.__mro__)
print('MultiplyTool bases:', MultiplyTool.__mro__)
print('BaseTool:', BaseTool)
print('MultiplyInput source:')
print(inspect.getsource(MultiplyInput))
```

```python
multiply_tool = MultiplyTool()
result = multiply_tool.invoke({'a':5,'b':6})
print(result)
```
